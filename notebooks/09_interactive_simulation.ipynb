## 1. Live Interactive Widget Simulation
Drag the sliders to change parameters and observe how feedback delays ($\tau$) and response speeds ($\lambda$) alter oscillation stability.


```python
def run_interactive(alpha, lambda_, tau, sigma, theta, nu):
    sim = PendulumSimulation(
        alpha=alpha, lambda_=lambda_, tau=int(tau), 
        sigma=sigma, theta=theta, nu=nu
    )
    sim.run(steps=80)
    fig = sim.plot_time_series()
    plt.show()

interact(
    run_interactive,
    alpha=widgets.FloatSlider(value=0.15, min=0.01, max=0.5, step=0.01),
    lambda_=widgets.FloatSlider(value=0.2, min=0.01, max=0.9, step=0.01),
    tau=widgets.IntSlider(value=2, min=0, max=6, step=1),
    sigma=widgets.FloatSlider(value=0.4, min=0.01, max=1.0, step=0.05),
    theta=widgets.FloatSlider(value=1.0, min=0.1, max=3.0, step=0.1),
    nu=widgets.FloatSlider(value=0.08, min=0.0, max=0.3, step=0.01)
);
```
